```
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis   import LinearDiscriminantAnalysis
```

```
iris = datasets.load_iris()
x = iris.data
y = iris.target
targets_name = iris.target_names
print(targets_name)
pca = PCA(n_components=2)
x_r = pca.fit(x).transform(x)

lda = LinearDiscriminantAnalysis(n_components=2)
x_r2 = lda.fit(x,y).transform(x)
print(x_r2)
print('explained variance rasio(first two component):%s' %str(pca.explained_variance_ratio_))
```

```
plt.figure()
colors = ['navy','green','red']
lw=2
for color,i,target_name in zip(colors,[0,1,2],targets_name):
  plt.scatter(x_r[y==i,0],x_r[y==i,1],color=color,lw=lw,label=target_name)
plt.legend(loc='best',shadow=False,scatterpoints=1)
plt.title('plt pca')

plt.figure()
colors = ['navy','green','red']
lw=2
for color,i,target_name in zip(colors,[0,1,2],targets_name):
  plt.scatter(x_r2[y==i,0],x_r2[y==i,1],color=color,lw=lw,label=target_name)
plt.legend(loc='best',shadow=False,scatterpoints=1)
plt.title('plt lda')
```
